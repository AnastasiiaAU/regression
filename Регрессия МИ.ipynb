```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
import numpy as np
```

```python
# считываем файл и выводим первые 5 строк
file = pd.read_excel("Ecomm research results.xlsx")
file.head()
```

```python
# создание отдельной таблицы с избанными факторами + корреляционная матрица 
factors1 = file[["how_often", "feature_fast_search", "easy_to_comparing_offers","easy_to_purchase", "saving_money",
              "saving_time", "friendly_attitude_to_customers", "professionalism_of_the_employees", 
               "trust_information", "proucts_not_overpriced", "eshops_are_honest", "eshops_respect_interests_of_consumers",
                'not_overpricing_when_selling', "not_cheating_consumers", "comply_with_obligations", "can_trust_eshops"]]
```

```python
del factors1['can_trust_eshops']
del factors1['easy_to_comparing_offers']
factors1.corr()
```

```python
import statsmodels.formula.api as smf
# построение регрессионной модели по избранным факторам
lm = smf.ols(formula='how_often ~ feature_fast_search + easy_to_purchase + saving_money + saving_time + friendly_attitude_to_customers + professionalism_of_the_employees + trust_information + proucts_not_overpriced + eshops_are_honest + eshops_respect_interests_of_consumers + not_overpricing_when_selling + not_cheating_consumers + comply_with_obligations', data=factors1).fit()

# вывод коэффициентов
lm.params
```

```python
lm.summary() # вывов саммари по модели
```
