```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
import numpy as np
```

```python
# считываем файл и выводим первые 5 строк
file = pd.read_excel("Ecomm research results.xlsx")
file.head()
```

```python
factors = file[["like_buy_in_internet", "trust_information", "feature_fast_search", 'easy_to_comparing_offers', 
                 'easy_to_purchase', 'saving_money', 'saving_time', 'friendly_attitude_to_customers', 
                 "professionalism_of_the_employees", 'increase_shopping_performance']]
factors.corr()
```

```python
import statsmodels.formula.api as smf
# построение регрессионной модели по избранным факторам
lm = smf.ols(formula='like_buy_in_internet ~ trust_information + feature_fast_search + easy_to_comparing_offers + easy_to_purchase + saving_money + saving_time + friendly_attitude_to_customers + professionalism_of_the_employees +increase_shopping_performance', data=file).fit()

# вывод коэффициентов
lm.params
```
